```python
import matplotlib.pyplot as plt
import numpy as np

import week3
```

```python
np.set_printoptions(precision=3)
```

```python
# Payoff matrix for the row player in the Rock-Paper-Scissors game
matrix = np.array([[0, -1, 1], [1, 0, -1], [-1, 1, 0]], dtype=np.float32)

# Strategies for the players
row_strategy = np.array([0.1, 0.2, 0.7], dtype=np.float32)
col_strategy = np.array([0.3, 0.2, 0.5], dtype=np.float32)
```

```python
row_delta, col_delta = week3.compute_deltas(row_strategy, col_strategy, matrix, -matrix)
print(f'Deltas: {row_delta:.3f}, {col_delta:.3f}')
```

```python
exploitability = week3.compute_exploitability(row_strategy, col_strategy, matrix, -matrix)
print(f'Exploitability: {exploitability:.3f}')
```

```python
epsilon = week3.compute_epsilon(row_strategy, col_strategy, matrix, -matrix)
print(f'Epsilon: {epsilon:.3f}')
```

```python
matrix = np.array([[1, 2, 3], [-1, -2, -1], [3, 1, -4]], dtype=np.float32)

strategies = week3.fictitious_play(matrix, -matrix, 100, naive=True)
exploitability_values = week3.plot_exploitability(matrix, -matrix, strategies, label='Naive Fictitious Play')

print(f'Final exploitability value when naive=True: {exploitability_values[-1]}', end='\n\n')

strategies = week3.fictitious_play(matrix, -matrix, 100, naive=False)
exploitability_values = week3.plot_exploitability(matrix, -matrix, strategies, label='Fictitious Play')

print(f'Final exploitability value naive=False: {exploitability_values[-1]}')

plt.legend()
plt.show()
```
