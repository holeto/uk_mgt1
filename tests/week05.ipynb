```python
import matplotlib.pyplot as plt
import numpy as np

import week3
import week5
```

```python
np.set_printoptions(precision=3)
```

```python
games = [
    (np.array([[1, 2, 3], [-1, -2, -1], [3, 1, -4]], dtype=np.float32), np.array([[-1, -2, -3], [1, 2, 1], [-3, -1, 4]], dtype=np.float32)),
    (np.array([[0, -1, 1], [1, 0, -1], [-1, 1, 0]], dtype=np.float32), np.array([[0, 1, -1], [-1, 0, 1], [1, -1, 0]], dtype=np.float32)),
    (np.array([[-1, -3], [0, -2]], dtype=np.float32), np.array([[-1, 0], [-3, -2]], dtype=np.float32)),
    (np.array([[0, -1], [1, -10]], dtype=np.float32), np.array([[0, 1], [-1, -10]], dtype=np.float32)),
    (np.array([[10, 0], [0, 7]], dtype=np.float32), np.array([[7, 0], [0, 10]], dtype=np.float32))
]
```

```python
for (matrix1, matrix2) in games:
    strategies = week3.fictitious_play(matrix1, matrix2, epochs=100, naive=False)
    exploitability_values = week3.plot_exploitability(matrix1, matrix2, strategies, label='FSP')

    print(f'Final exploitability value of FSP: {exploitability_values[-1]}', end='\n\n')

    strategies = week5.regret_minimization(matrix1, matrix2, epochs=100)
    exploitability_values = week3.plot_exploitability(matrix1, matrix2, strategies, label='Regret Minimization')

    print(f'Final exploitability value of RM: {exploitability_values[-1]}')

    plt.legend()
    plt.show()
    print()
```
