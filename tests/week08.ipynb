```python
import matplotlib.pyplot as plt
import numpy as np

import efg
import week6
import week8
```

```python
np.set_printoptions(precision=5)
```

```python
# In all examples, both Fictitious Play and CFR are initialized using uniform strategies
# You can compare the final strategies with optimal strategies listed on Wikipedia
# https://en.wikipedia.org/wiki/Kuhn_poker
kuhn_poker = efg.KuhnPoker()

strategies = week6.fictitious_play(kuhn_poker, epochs=2500, report=False)
exploitability_values = week6.plot_exploitability(kuhn_poker, strategies, 'FP')
print(f'Final exploitability value of FP: {exploitability_values[-1]:.5f}', end='\n\n')

strategies = week8.cfr(kuhn_poker, epochs=2500)
exploitability_values = week6.plot_exploitability(kuhn_poker, strategies, 'CFR')
print(f'Final exploitability value of CFR: {exploitability_values[-1]:.5f}', end='\n\n')

strategies = week8.cfr(kuhn_poker, epochs=2500, naive=True)
exploitability_values = week6.plot_exploitability(kuhn_poker, strategies, 'Naive CFR')
print(f'Final exploitability value of Naive CFR: {exploitability_values[-1]:.5f}')

plt.legend()
plt.show()
```

```python
# This implements the Poker-like game tree presented in the lecture
lecture_poker = efg.LecturePoker()

strategies = week6.fictitious_play(lecture_poker, epochs=1000, report=False)
exploitability_values = week6.plot_exploitability(lecture_poker, strategies, 'FP')
print(f'Final exploitability value of FP: {exploitability_values[-1]:.5f}', end='\n\n')

strategies = week8.cfr(lecture_poker, epochs=1000)
exploitability_values = week6.plot_exploitability(lecture_poker, strategies, 'CFR')
print(f'Final exploitability value of CFR: {exploitability_values[-1]:.5f}', end='\n\n')

strategies = week8.cfr(lecture_poker, epochs=1000, naive=True)
exploitability_values = week6.plot_exploitability(lecture_poker, strategies, 'Naive CFR')
print(f'Final exploitability value of Naive CFR: {exploitability_values[-1]:.5f}')

plt.legend()
plt.show()
```
