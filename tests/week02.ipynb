```python
# Taken from the slides for Week 2
matrix1 = np.array([[0, 0, -10], [1, -10, -10], [-10, -10, -10]], dtype=np.float32)
matrix2 = np.array([[0, 1, -10], [0, -10, -10], [-10, -10, -10]], dtype=np.float32)

col_strategy = week2.verify_support_one_side([0, 1], [0, 1, 2], matrix1)
np.testing.assert_allclose(col_strategy, [0.90909, 0.09090, 0.0], rtol=1e-3)

row_strategy = week2.verify_support_one_side([0, 1, 2], [0, 1], np.transpose(matrix2))
assert row_strategy is None, 'Wrong result!'
```

```python
# Prisoners' Dilemma
matrix1 = np.array([[-1, -3], [0, -2]], dtype=np.float32)
matrix2 = np.array([[-1, 0], [-3, -2]], dtype=np.float32)

equilibria = week2.support_enumeration(matrix1, matrix2)
for (row_strategy, col_strategy) in equilibria:
    row_strategy = np.asarray(row_strategy)
    col_strategy = np.asarray(col_strategy)

    print(f'Row strategy: {row_strategy}, Col strategy: {col_strategy}')
```

```python
# Game of Chicken
matrix1 = np.array([[0, -1], [1, -10]], dtype=np.float32)
matrix2 = np.array([[0, 1], [-1, -10]], dtype=np.float32)

equilibria = week2.support_enumeration(matrix1, matrix2)
for (row_strategy, col_strategy) in equilibria:
    row_strategy = np.asarray(row_strategy)
    col_strategy = np.asarray(col_strategy)

    print(f'Row strategy: {row_strategy}, Col strategy: {col_strategy}')
```
